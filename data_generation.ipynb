```python
import numpy as np
item_counts = 1

for index, item in items_meta_data.iterrows():
    
    if index < 3213:
        continue
    
    filename = 'train_data_by_item/item_' + str(item['item_nbr']) + '.csv'
    
    # load the train data of each item
    train_data = pd.read_csv(filename, parse_dates = ['date'], dayfirst = True, index_col = 'date')
    print(train_data.shape)
    
    if train_data.empty:
        print('Item %d doesnt exists...' % item['item_nbr'])
        continue
    
    # select a time horizon
    start_date = '2017-01-01'
    end_date = '2017-08-15'
    data = train_data.loc[start_date:end_date]
    
    
    if data.empty:
        print('No data exists for the date range...')
        continue
    
    # indexing the loaded data
    data['day'] = data.index.day
    data['weekday'] = data.index.weekday
    data['month'] = data.index.month
    data['dayofyear'] = data.index.dayofyear # The ordinal day of the year
    
    print(data.shape)
    
    
    # create date-store dataframe
    column_names = ['s%d' % c for c in range(1, 55)]
    Xtrain = pd.DataFrame(0.0, index = np.arange(227), columns = column_names)
    
    #X = X.reset_index()
    #Xtrain.t.values[X.index.dayofyear] = np.log(X.unit_sales.values)
    for store_i in range(1,55):
        try:
            X = data.groupby(['store_nbr']).get_group(store_i) 
            store_name = 's' + str(store_i)
            
            # if unit sales are negative
            X['unit_sales'].values[X['unit_sales'] < 0.0 ] = 0.0
            
            
            # indices are different for array and dayofyear
            #Xtrain[store_name].values[X.index.dayofyear-1] = np.log(X.unit_sales.values)
            Xtrain[store_name].values[X.index.dayofyear-1] = X.unit_sales.values
        except KeyError:
            print('Store %d not found' %store_i)
    
    # store in train_data_unit_sales folder
    Xtrain.to_csv('train_data_unit_sales/items_' + str(item['item_nbr']) + '.csv', index = False)
    print('train data has been created for item: ' + str(item['item_nbr']))
    
    #if item_counts == 4100:
    #    break
    #item_counts += 1
    
    
print("Finish creating train data....")
```
